## Introduction

The dataset under analysis contains information about location, salary, gender, and other features. The goal of this exploratory data analysis (EDA) is to investigate the relationships between various features, with a focus on salary as the target variable. We aim to test specific claims related to salary, explore relationships between gender and specialization, and identify other useful insights that can help in understanding the patterns in the data.

The dataset is a mix of numerical and categorical variables, requiring various univariate and bivariate analysis techniques to uncover trends and correlations.

## Histograms

```python
def hist(i,c):
    plt.figure(figsize=(10, 5))
    sns.histplot(dt[i], bins=10, color=c, alpha=0.9)
    plt.title(f'Distribution of {i}', fontsize=17)
    plt.ylabel('Frequency', fontsize=12)
    plt.show()
```

```python
def count1(i,w,h,c,r,n):
    data = dt[dt[i].isin(dt[i].value_counts().nlargest(n).index)]
    plt.figure(figsize=(w, h))
    ax = sns.countplot(x= data[i], alpha=0.9, palette='viridis')
    plt.title(f'top {n} {i}', fontsize=17)
    plt.xticks(rotation=r)
    plt.xlabel(i)
    plt.ylabel('Frequency')
```

## pie plots

```python
def barplot_(i,j, w, h, r, n):
    plt.figure(figsize=(15,5))
    top_10_values = df[i].value_counts().nlargest(n).index
    d = df[df[i].isin(top_10_values)]
    sns.barplot(x=d[i],y=d[j],data=dt)
    plt.xlabel(i)
    plt.ylabel(j)
    plt.xticks(rotation=90)
    plt.title(f'{i} vs {j}')
    sns.despine() 
    plt.show()
```

```python
def swarmplot_(i, j, n):
    plt.figure(figsize=(15, 5))
    top_n_values = df[i].value_counts().nlargest(n).index
    d = df[df[i].isin(top_n_values)]
    sns.swarmplot(x=d[i], y=d[j], data=d)
    plt.xlabel(i)
    plt.ylabel(j)
    plt.xticks(rotation=90)
    plt.title(f'{i} vs {j}')
    sns.despine() 
    plt.show()
```
